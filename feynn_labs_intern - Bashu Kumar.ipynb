```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
scaler = StandardScaler()
data_x_standardized = scaler.fit_transform(data_x)

# Perform PCA
pca = PCA()
data_pca = pca.fit(data_x_standardized)

# Print summary
print("Explained Variance Ratio:")
print(data_pca.explained_variance_ratio_)
print("\nSummary of PCA:")
print(pd.DataFrame({'Standard deviation': np.sqrt(data_pca.explained_variance_),'Proportion of Variance': data_pca.explained_variance_ratio_,
                    'Cumulative Proportion': np.cumsum(data_pca.explained_variance_ratio_)}))
```

```python
# Assuming data_pca is your PCA object from the previous example
# Replace it with the actual name of your PCA object if different

# Print standard deviations
print("Standard deviations (1, .., p=11):")
print(np.round(np.sqrt(data_pca.explained_variance_), 1))

# Print rotation matrix
print("\nRotation (n x k) = (11 x 11):")
rotation_matrix = pd.DataFrame(data_pca.components_.T, columns=[f'PC{i+1}' for i in range(data_pca.components_.shape[0])])
print(rotation_matrix)
```

```python
import matplotlib.pyplot as plt

# Assuming data_pca is your PCA object from the previous example
# Replace it with the actual name of your PCA object if different

# Transform the data using the first two principal components
data_pca_transformed = data_pca.transform(data_x_standardized)

# Plot the transformed data
plt.scatter(data_pca_transformed[:, 0], data_pca_transformed[:, 1], color='grey')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Projection')
plt.show()

# Plot projected axes
plt.figure()
plt.plot(data_pca.components_[0, :], label='PC1')
plt.plot(data_pca.components_[1, :], label='PC2')
plt.xlabel('Original Features')
plt.ylabel('Projection Weights')
plt.title('Projected Axes')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Assuming data_x is your binary matrix from the previous example
# Replace it with the actual name of your binary matrix if different

# Set seed for reproducibility
np.random.seed(1234)

# Perform k-means clustering for k=2 to k=8 with 10 repetitions
results = []
for k in range(2, 9):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=1234)
    kmeans.fit(data_x)
    results.append(kmeans)

# Choose the best result based on inertia or other criteria
# You may want to customize the selection based on your specific criteria
best_kmeans = results[3]  # For example, choosing k=5

# Relabel the clusters
labels = best_kmeans.labels_

# Print the cluster assignments
print(labels)
```

```python
import matplotlib.pyplot as plt

# Assuming best_kmeans is your selected k-means result from the previous example
# Replace it with the actual name of your k-means result if different

# Plot the clustering results for each k
inertia_values = [kmeans.inertia_ for kmeans in results]

plt.plot(range(2, 9), inertia_values, marker='o')
plt.xlabel('Number of Segments (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

## Selecting target segment

```python
#Calculating the mean
#Visit frequency
data['VisitFrequency'] = LabelEncoder().fit_transform(data['VisitFrequency'])
visit = data.groupby('cluster_num')['VisitFrequency'].mean()
visit = visit.to_frame().reset_index()
visit
```

```python
#Like
data['Like'] = LabelEncoder().fit_transform(data['Like'])
Like = data.groupby('cluster_num')['Like'].mean()
Like = Like.to_frame().reset_index()
Like
```

```python
#Gender
data['Gender'] = LabelEncoder().fit_transform(data['Gender'])
Gender = data.groupby('cluster_num')['Gender'].mean()
Gender = Gender.to_frame().reset_index()
Gender
```

```python
segment = Gender.merge(Like, on='cluster_num', how='left').merge(visit, on='cluster_num', how='left')
segment
```

## Step 8: Selecting (the) Target Segment(s)

```python
#Target segments

plt.figure(figsize = (9,4))
sns.scatterplot(x = "VisitFrequency", y = "Like",data=segment,s=400, color="r")
plt.title("Simple segment evaluation plot for the fast food data set",
          fontsize = 15)
plt.xlabel("Visit", fontsize = 12)
plt.ylabel("Like", fontsize = 12)
plt.grid()
plt.show()
```
